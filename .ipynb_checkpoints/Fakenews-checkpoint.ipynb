```python
import re
```

```python
def wordopt(text):
  #Convert into lowercase
  text = text.lower()

  #Remove URLs
  text = re.sub('https?://\S+|www\.\S+', '', text)

  #Remove HTMLs
  text = re.sub('<.*?>+', '', text)

  #Remove Punctuation
  text = re.sub('[%s]' % re.escape(string.punctuation), '', text)

  #Remove digits
  text = re.sub('\w*\d\w*', '', text)

  #Remove newline characters
  text = re.sub('\n', '', text)

  return text
```
