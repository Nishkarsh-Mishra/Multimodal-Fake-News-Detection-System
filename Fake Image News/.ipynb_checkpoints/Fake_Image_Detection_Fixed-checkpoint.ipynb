### Import Libraries

```python
import os
import numpy as np
import matplotlib.pyplot as plt
import cv2

from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, accuracy_score

import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.applications.mobilenet_v2 import preprocess_input  # ✅ FIX: correct preprocessor
```

### Load Data using ImageDataGenerator

**BUG FIXED:** MobileNetV2 requires `preprocess_input` (scales pixels to [-1, 1]).
Using `rescale=1./255` was the main cause of low accuracy (~61%).

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications.mobilenet_v2 import preprocess_input

IMG_SIZE = 224
BATCH_SIZE = 32

# ✅ FIX 1: Use preprocess_input instead of rescale=1./255
# ✅ FIX 2: Separate train and val datagens — val should NOT have augmentation
# ✅ IMPROVEMENT: Added stronger augmentation (brightness, shear) for fake image detection
train_datagen = ImageDataGenerator(
    preprocessing_function=preprocess_input,   # ✅ Correct preprocessing for MobileNetV2
    validation_split=0.2,
    rotation_range=20,
    zoom_range=0.2,
    horizontal_flip=True,
    shear_range=0.1,                            # ✅ Added: helps with perspective distortions
    brightness_range=[0.8, 1.2],               # ✅ Added: helps detect lighting artefacts in fakes
    width_shift_range=0.1,
    height_shift_range=0.1,
)

# ✅ FIX 3: Validation datagen — only preprocessing, NO augmentation
val_datagen = ImageDataGenerator(
    preprocessing_function=preprocess_input,
    validation_split=0.2
)

train_data = train_datagen.flow_from_directory(
    "IMAGES",
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode='binary',
    subset='training',
    shuffle=True
)

val_data = val_datagen.flow_from_directory(
    "IMAGES",
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode='binary',
    subset='validation',
    shuffle=False  # ✅ Keep False for evaluation consistency
)

print(train_data.class_indices)
print(f"Training samples: {train_data.samples}")
print(f"Validation samples: {val_data.samples}")
```

### Build Model (Transfer Learning + Fine-Tuning)

**Improvements:**
- Added a second Dense layer (256 units) for better feature extraction
- Deeper Dropout to reduce overfitting
- Lower learning rate (`1e-4`) suited for transfer learning
- Fine-tuning enabled for the last 30 layers of MobileNetV2

```python
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras import layers, models
from tensorflow.keras.optimizers import Adam

base_model = MobileNetV2(
    weights='imagenet',
    include_top=False,
    input_shape=(224, 224, 3)
)

# ✅ FIX 4: Freeze all layers first, then unfreeze last 30 for fine-tuning
base_model.trainable = False

# Unfreeze last 30 layers for fine-tuning
for layer in base_model.layers[-30:]:
    if not isinstance(layer, layers.BatchNormalization):  # Keep BN frozen
        layer.trainable = True

# ✅ IMPROVEMENT: Deeper classifier head
x = base_model.output
x = layers.GlobalAveragePooling2D()(x)
x = layers.BatchNormalization()(x)
x = layers.Dense(256, activation='relu')(x)   # ✅ Added wider first dense layer
x = layers.Dropout(0.4)(x)
x = layers.Dense(128, activation='relu')(x)
x = layers.Dropout(0.3)(x)
output = layers.Dense(1, activation='sigmoid')(x)

model = models.Model(inputs=base_model.input, outputs=output)

# ✅ FIX 5: Lower learning rate for transfer learning (1e-4 instead of default 1e-3)
model.compile(
    optimizer=Adam(learning_rate=1e-4),
    loss='binary_crossentropy',
    metrics=['accuracy']
)

trainable_count = sum(1 for l in model.layers if l.trainable)
print(f"Trainable layers: {trainable_count} / {len(model.layers)}")
model.summary()
```

### Train Model

```python
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint

early_stop = EarlyStopping(
    monitor='val_accuracy',   # ✅ Monitor accuracy, not just loss
    patience=5,               # ✅ More patience to allow fine-tuning to converge
    restore_best_weights=True
)

reduce_lr = ReduceLROnPlateau(
    monitor='val_loss',
    factor=0.3,
    patience=3,
    min_lr=1e-7,
    verbose=1
)

# ✅ IMPROVEMENT: Save the best model automatically
checkpoint = ModelCheckpoint(
    'best_model.keras',
    monitor='val_accuracy',
    save_best_only=True,
    verbose=1
)

history = model.fit(
    train_data,
    validation_data=val_data,
    epochs=30,           # ✅ More epochs; early stopping will prevent over-training
    callbacks=[early_stop, reduce_lr, checkpoint]
)
```

### Plot Training History

```python
import matplotlib.pyplot as plt

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

ax1.plot(history.history['accuracy'], label='Train Accuracy')
ax1.plot(history.history['val_accuracy'], label='Val Accuracy')
ax1.set_title('Model Accuracy')
ax1.set_xlabel('Epoch')
ax1.set_ylabel('Accuracy')
ax1.legend()
ax1.grid(True)

ax2.plot(history.history['loss'], label='Train Loss')
ax2.plot(history.history['val_loss'], label='Val Loss')
ax2.set_title('Model Loss')
ax2.set_xlabel('Epoch')
ax2.set_ylabel('Loss')
ax2.legend()
ax2.grid(True)

plt.tight_layout()
plt.show()
```

### Evaluate Model

```python
loss, accuracy = model.evaluate(val_data)
print(f"Validation Accuracy: {accuracy:.4f}")
print(f"Validation Loss: {loss:.4f}")

# ✅ IMPROVEMENT: Detailed classification report
val_data.reset()
y_pred_probs = model.predict(val_data)
y_pred = (y_pred_probs > 0.5).astype(int).flatten()
y_true = val_data.classes

print("\nClassification Report:")
print(classification_report(y_true, y_pred, target_names=list(val_data.class_indices.keys())))
```

### Test on New Image

**BUG FIXED:** Original code used BGR from cv2 without converting to RGB.
MobileNetV2's `preprocess_input` expects RGB images.

```python
import cv2
import numpy as np
from tensorflow.keras.applications.mobilenet_v2 import preprocess_input

def predict_image(path):
    img = cv2.imread(path)

    if img is None:
        print("❌ Error: Image not found or unsupported format")
        return

    # ✅ FIX 7: Convert BGR (cv2 default) to RGB
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    img = cv2.resize(img, (224, 224))

    # ✅ FIX 8: Use preprocess_input instead of /255.0
    img = preprocess_input(img.astype(np.float32))
    img = np.reshape(img, (1, 224, 224, 3))

    prediction = model.predict(img)[0][0]

    print(f"Prediction Score: {prediction:.4f}")

    if prediction > 0.5:
        print(f"✅ Real Image  (confidence: {prediction*100:.1f}%)")
    else:
        print(f"❌ Fake Image  (confidence: {(1-prediction)*100:.1f}%)")
```

### FINAL MINI UI CODE (Jupyter)

**BUG FIXED:** `file_info['content']` returns memoryview in newer ipywidgets — call `.tobytes()` on it.

```python
import ipywidgets as widgets
from IPython.display import display, clear_output
import numpy as np
from PIL import Image
import io
from tensorflow.keras.applications.mobilenet_v2 import preprocess_input

# Upload widget
upload = widgets.FileUpload(
    accept='.jpg,.jpeg,.png',
    multiple=False
)

# Button
button = widgets.Button(description="Predict Image", button_style='primary')

# Output area
output = widgets.Output()

def on_button_click(b):
    with output:
        clear_output()

        if not upload.value:
            print("❌ Please upload an image first")
            return

        file_info = upload.value[0]

        # ✅ FIX 9: Handle both bytes and memoryview from ipywidgets
        content = file_info['content']
        if hasattr(content, 'tobytes'):
            content = content.tobytes()

        # Convert to PIL image
        image = Image.open(io.BytesIO(content)).convert('RGB')  # ✅ Always convert to RGB
        display(image)  # Preview

        # Resize & preprocess
        img = image.resize((224, 224))
        img_array = np.array(img, dtype=np.float32)

        # ✅ FIX 10: Use correct preprocess_input
        img_array = preprocess_input(img_array)
        img_array = np.reshape(img_array, (1, 224, 224, 3))

        # Prediction
        prediction = model.predict(img_array)[0][0]

        print(f"\nPrediction Score: {prediction:.4f}")

        if prediction > 0.5:
            print(f"✅ Real Image  (confidence: {prediction*100:.1f}%)")
        else:
            print(f"❌ Fake Image  (confidence: {(1-prediction)*100:.1f}%)")

# Button click event
button.on_click(on_button_click)

# Display UI
display(widgets.VBox([upload, button, output]))
```
