```python
import re
```

 ### Define Text Preprocessing Function (wordopt)

```python
def wordopt(text):
  #Convert into lowercase
  text = text.lower()

  #Remove URLs
  text = re.sub('https?://\S+|www\.\S+', '', text)

  #Remove HTMLs
  text = re.sub('<.*?>+', '', text)

  #Remove Punctuation
  text = re.sub('[%s]' % re.escape(string.punctuation), '', text)

  #Remove digits
  text = re.sub('\w*\d\w*', '', text)

  #Remove newline characters
  text = re.sub('\n', '', text)

  return text
```

```python
import string
combined_df['text'] = combined_df['text'].apply(wordopt)
```

```python
x = combined_df['text']
y = combined_df['label']
```

```python
x
```

```python
y
```

### Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3, random_state=42)
```

```python
x_train.shape
```

```python
x_test.shape
```

### TF-IDF Vectorization

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorization = TfidfVectorizer
```

```python
vectorizer_instance = vectorization()
xv_train = vectorizer_instance.fit_transform(x_train)
xv_test = vectorizer_instance.transform(x_test)
```

```python
xv_train
```

```python
xv_test
```

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

LR = LogisticRegression()
LR.fit(xv_train, y_train)
```

```python
pred_lr = LR.predict(xv_test)
```

```python
LR.score(xv_test, y_test)
```

### Classification Report

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, pred_lr))
```

### Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier

DTC = DecisionTreeClassifier()
DTC.fit(xv_train, y_train)
```

```python
pred_dtc = DTC.predict(xv_test)
```

```python
DTC.score(xv_test, y_test)
```

```python
print('Classification Report for Decision Tree Classifier')
print(classification_report(y_test, pred_dtc))
```

### Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc= RandomForestClassifier(random_state=0)
rfc.fit(xv_train, y_train)
```

```python
predict_rfc = rfc.predict(xv_test)
```

```python
rfc.score(xv_test, y_test)
```

```python
print(classification_report(y_test,predict_rfc))
```

### Gradient Boosting Classifier

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
gbc = GradientBoostingClassifier(random_state=0)
gbc.fit(xv_train, y_train)
```

```python
gbc.score(xv_test, y_test)
```

```python
print(classification_report(y_test,gbc.predict(xv_test)))
```

### Xgboost Classifier

```python
import xgboost as xgb

xgb_model = xgb.XGBClassifier(random_state=0, eval_metric='logloss')
xgb_model.fit(xv_train, y_train)

# Make predictions
pred_xgb = xgb_model.predict(xv_test)

# Evaluate the model
print(f"XGBoost Accuracy: {xgb_model.score(xv_test, y_test)}")
```

```python
from sklearn.metrics import classification_report

print("Classification Report for XGBoost:")
print(classification_report(y_test, pred_xgb))
```

```python
def output_label(n):
  if n==0:
    return "It is a Fake News"
  elif n==1:
    return "It is a Genuine News"
```

```python
def manual_testing(news):
  testing_news = {"text":[news]}
  news_def_test = pd.DataFrame(testing_news)
  news_def_test["text"] = news_def_test["text"].apply(wordopt)
  new_x_test = news_def_test["text"]
  new_xv_test = vectorizer_instance.transform(new_x_test)

  pred_lr = LR.predict(new_xv_test)
  pred_xgb = xgb_model.predict(new_xv_test)
  pred_dtc = DTC.predict(new_xv_test)
  pred_gbc = gbc.predict(new_xv_test)
  pred_rfc = rfc.predict(new_xv_test)

  # Collect all predictions
  all_predictions = [pred_lr[0], pred_xgb[0], pred_dtc[0], pred_gbc[0], pred_rfc[0]]

  # Implement majority voting
  # If sum of predictions for '1' is >= 3, then it's a genuine news (1), otherwise fake news (0)
  majority_vote = 1 if sum(all_predictions) >= 3 else 0

  return 'Majority Vote Prediction: {}'.format(output_label(majority_vote))
```

```python
news_article = str(input())
```

```python
print(manual_testing(news_article))
```
