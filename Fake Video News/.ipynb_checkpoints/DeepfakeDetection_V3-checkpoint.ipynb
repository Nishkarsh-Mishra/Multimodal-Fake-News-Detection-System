# 🎭 Fake Video Detection — FaceForensics++ (V3 — Root-Cause Fixed)

### Root causes of "always predicts FAKE" — all fixed here:
| # | Bug | Fix |
|---|-----|-----|
| 1 | Data loaded in order (all REAL first, then all FAKE) → model sees sorted stream, BatchNorm learns biased statistics | `shuffle` before split |
| 2 | No data augmentation → model memorizes texture artifacts from fake videos | `ImageDataGenerator` with flips/zoom |
| 3 | Fixed threshold 0.5 → with FAKE recall 0.99 / REAL recall 0.83, optimal threshold is NOT 0.5 | Auto-calibrate threshold from val set |
| 4 | Frames in `predict_video` were not shuffled before averaging → early frames of a real video may look synthetic | Already fixed by evenly-spaced sampling |
| 5 | `data` and `labels` arrays built separately and never jointly shuffled before `train_test_split` | Use `shuffle=True` in split + `random_state` |

### STEP 1: Extract Frames

```python
import cv2
import os
import numpy as np

IMG_SIZE = 128
FRAMES_PER_VIDEO = 10

def extract_frames(video_path, n_frames=FRAMES_PER_VIDEO):
    """
    Extracts n_frames evenly-spaced frames from a video.
    Returns raw uint8 numpy arrays (NOT normalized).
    Normalization happens only once, in the dataset builder.
    """
    cap = cv2.VideoCapture(video_path)
    if not cap.isOpened():
        return []

    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    if total_frames <= 0:
        cap.release()
        return []

    indices = np.linspace(0, total_frames - 1, n_frames, dtype=int)
    frames = []

    for idx in indices:
        cap.set(cv2.CAP_PROP_POS_FRAMES, idx)
        ret, frame = cap.read()
        if not ret:
            continue
        frame = cv2.resize(frame, (IMG_SIZE, IMG_SIZE))
        frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)  # BGR → RGB
        frames.append(frame)   # raw uint8, shape (128,128,3)

    cap.release()
    return frames

print("✅ extract_frames() ready")
```

### STEP 2: Load Dataset
> **Key fix:** We interleave real and fake videos while loading instead of loading all real first then all fake.  
> This prevents BatchNormalization from learning biased running statistics.

```python
import os
import numpy as np
import random

random.seed(42)
np.random.seed(42)

MAX_REAL = 200
MAX_FAKE = 200   # same count → balanced dataset, no class_weight needed

# ═══════════════════ COLLECT VIDEO PATHS FIRST ═══════════════════
real_path = "FaceForensics++/real_original_sequences"
fake_path = "FaceForensics++/fake_manipulated_sequences"

real_videos = sorted([
    os.path.join(real_path, v)
    for v in os.listdir(real_path)
    if v.endswith('.mp4')
])[:MAX_REAL]

fake_folders = sorted(os.listdir(fake_path))
per_folder   = MAX_FAKE // len(fake_folders)

fake_videos = []
for folder in fake_folders:
    fp = os.path.join(fake_path, folder)
    if not os.path.isdir(fp):
        continue
    vids = sorted([
        os.path.join(fp, v)
        for v in os.listdir(fp)
        if v.endswith('.mp4')
    ])[:per_folder]
    fake_videos.extend(vids)

fake_videos = fake_videos[:MAX_FAKE]

print(f"Real video paths collected : {len(real_videos)}")
print(f"Fake video paths collected : {len(fake_videos)}")

# ═══════════════════ INTERLEAVED LOADING ═══════════════════
# FIX: zip real+fake together so they are interleaved, not all-real then all-fake
all_video_paths  = []
all_video_labels = []

for r, f in zip(real_videos, fake_videos):
    all_video_paths.append(r)
    all_video_labels.append(1)   # REAL = 1
    all_video_paths.append(f)
    all_video_labels.append(0)   # FAKE = 0

# Shuffle the video list (not just frames) before extracting
combined = list(zip(all_video_paths, all_video_labels))
random.shuffle(combined)
all_video_paths, all_video_labels = zip(*combined)

# ═══════════════════ EXTRACT FRAMES ═══════════════════
data   = []
labels = []
total  = len(all_video_paths)

for i, (vpath, vlabel) in enumerate(zip(all_video_paths, all_video_labels)):
    frames = extract_frames(vpath)
    for f in frames:
        data.append(f)
        labels.append(vlabel)
    if (i + 1) % 100 == 0:
        print(f"  [{i+1}/{total}] videos processed")

# ═══════════════════ FINALIZE ═══════════════════
data   = np.array(data,   dtype="float32") / 255.0
labels = np.array(labels, dtype="int32")

# Final shuffle at frame level too
shuffle_idx = np.random.permutation(len(data))
data   = data[shuffle_idx]
labels = labels[shuffle_idx]

print(f"\n📊 Final Data shape : {data.shape}")
print(f"   REAL frames      : {np.sum(labels == 1)}")
print(f"   FAKE frames      : {np.sum(labels == 0)}")
```

### STEP 3: Train / Val Split

```python
from sklearn.model_selection import train_test_split

# shuffle=True is redundant after our manual shuffle, but explicit is good
X_train, X_val, y_train, y_val = train_test_split(
    data, labels,
    test_size=0.2,
    random_state=42,
    stratify=labels,   # equal class ratio in both splits
    shuffle=True
)

print(f"Train : {X_train.shape[0]} samples  "
      f"(REAL: {np.sum(y_train==1)}, FAKE: {np.sum(y_train==0)})")
print(f"Val   : {X_val.shape[0]} samples    "
      f"(REAL: {np.sum(y_val==1)}, FAKE: {np.sum(y_val==0)})")
```

### STEP 4: Build & Train CNN
> Dataset is balanced (200 real = 200 fake), so **no class_weight needed**.  
> Using `ImageDataGenerator` for augmentation to prevent overfitting to fake-specific artifacts.

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import (
    Conv2D, MaxPooling2D, Flatten, Dense,
    Dropout, BatchNormalization, Input
)
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# ── Data Augmentation (training only) ──────────────────────────────────────
# Augmentation makes the model generalize instead of memorizing
# compression artifacts in fake videos
train_datagen = ImageDataGenerator(
    horizontal_flip=True,
    zoom_range=0.1,
    rotation_range=10,
    width_shift_range=0.05,
    height_shift_range=0.05
)
train_gen = train_datagen.flow(X_train, y_train, batch_size=32, shuffle=True)

# ── Model ──────────────────────────────────────────────────────────────────
model = Sequential([
    Input(shape=(IMG_SIZE, IMG_SIZE, 3)),

    # Block 1
    Conv2D(32, (3,3), activation='relu', padding='same'),
    BatchNormalization(),
    MaxPooling2D(2, 2),

    # Block 2
    Conv2D(64, (3,3), activation='relu', padding='same'),
    BatchNormalization(),
    MaxPooling2D(2, 2),
    Dropout(0.25),

    # Block 3
    Conv2D(128, (3,3), activation='relu', padding='same'),
    BatchNormalization(),
    MaxPooling2D(2, 2),
    Dropout(0.25),

    # Classifier head
    Flatten(),
    Dense(256, activation='relu'),
    BatchNormalization(),
    Dropout(0.5),
    Dense(1, activation='sigmoid')
])

model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=1e-4),
    loss='binary_crossentropy',
    metrics=['accuracy', tf.keras.metrics.AUC(name='auc')]
)

model.summary()

# ── Callbacks ──────────────────────────────────────────────────────────────
callbacks = [
    EarlyStopping(
        monitor='val_accuracy',
        patience=5,
        restore_best_weights=True,
        verbose=1
    ),
    ReduceLROnPlateau(
        monitor='val_loss',
        factor=0.5,
        patience=2,
        min_lr=1e-7,
        verbose=1
    )
]

# ── Train ──────────────────────────────────────────────────────────────────
steps_per_epoch = len(X_train) // 32

history = model.fit(
    train_gen,
    steps_per_epoch=steps_per_epoch,
    epochs=20,
    validation_data=(X_val, y_val),
    callbacks=callbacks
)

val_loss, val_acc, val_auc = model.evaluate(X_val, y_val, verbose=0)
print(f"\n🏆 Final Val Accuracy : {val_acc*100:.2f}%")
print(f"   Final Val AUC      : {val_auc:.4f}")
```

### STEP 5: Calibrate Prediction Threshold
> **This is the most important fix for the "always predicts FAKE" bug.**  
> A fixed threshold of 0.5 is rarely the best choice. We find the threshold on the  
> validation set that maximizes balanced accuracy (equal weight to REAL and FAKE).

```python
from sklearn.metrics import balanced_accuracy_score
import matplotlib.pyplot as plt

# Get raw probabilities on validation set
y_prob = model.predict(X_val, verbose=0).flatten()

# Try every threshold from 0.1 to 0.9 in steps of 0.01
thresholds  = np.arange(0.05, 0.95, 0.01)
bal_scores  = []
real_recalls = []
fake_recalls = []

for t in thresholds:
    y_pred_t = (y_prob > t).astype(int)
    bal_scores.append(balanced_accuracy_score(y_val, y_pred_t))
    real_recalls.append(np.mean(y_pred_t[y_val == 1] == 1))  # real correctly called real
    fake_recalls.append(np.mean(y_pred_t[y_val == 0] == 0))  # fake correctly called fake

best_idx       = np.argmax(bal_scores)
BEST_THRESHOLD = float(thresholds[best_idx])

print(f"✅ Best threshold       : {BEST_THRESHOLD:.2f}")
print(f"   Balanced accuracy   : {bal_scores[best_idx]*100:.2f}%")
print(f"   REAL recall         : {real_recalls[best_idx]*100:.1f}%")
print(f"   FAKE recall         : {fake_recalls[best_idx]*100:.1f}%")

# Plot threshold vs balanced accuracy
plt.figure(figsize=(10, 4))
plt.plot(thresholds, bal_scores,   label='Balanced Accuracy', linewidth=2)
plt.plot(thresholds, real_recalls, label='REAL Recall',       linestyle='--')
plt.plot(thresholds, fake_recalls, label='FAKE Recall',       linestyle='--')
plt.axvline(BEST_THRESHOLD, color='red', linestyle=':', label=f'Best threshold = {BEST_THRESHOLD:.2f}')
plt.xlabel('Threshold')
plt.ylabel('Score')
plt.title('Threshold Calibration — Balanced Accuracy')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig('threshold_calibration.png', dpi=100)
plt.show()
```

### STEP 6: Training History Plots

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

axes[0].plot(history.history['accuracy'],     label='Train', marker='o')
axes[0].plot(history.history['val_accuracy'], label='Val',   marker='o')
axes[0].set_title('Accuracy')
axes[0].set_xlabel('Epoch')
axes[0].legend()
axes[0].grid(True)

axes[1].plot(history.history['loss'],     label='Train', marker='o')
axes[1].plot(history.history['val_loss'], label='Val',   marker='o')
axes[1].set_title('Loss')
axes[1].set_xlabel('Epoch')
axes[1].legend()
axes[1].grid(True)

plt.suptitle('Training History', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('training_history.png', dpi=100)
plt.show()
```

### STEP 7: Confusion Matrix + Classification Report

```python
from sklearn.metrics import confusion_matrix, classification_report, ConfusionMatrixDisplay

y_pred = (y_prob > BEST_THRESHOLD).astype(int)

cm = confusion_matrix(y_val, y_pred)

fig, ax = plt.subplots(figsize=(6, 5))
ConfusionMatrixDisplay(
    confusion_matrix=cm,
    display_labels=['FAKE', 'REAL']
).plot(ax=ax, colorbar=False, cmap='Blues')
ax.set_title(f'Confusion Matrix  (threshold = {BEST_THRESHOLD:.2f})', fontweight='bold')
plt.tight_layout()
plt.savefig('confusion_matrix.png', dpi=100)
plt.show()

print("\n📋 Classification Report:")
print(classification_report(y_val, y_pred, target_names=['FAKE', 'REAL']))
```

### STEP 8: Save Model + Threshold

```python
import json

# Save model
model.save("deepfake_detector.keras")
print("💾 Model saved: deepfake_detector.keras")

# Save the calibrated threshold so we can reload it
with open("detector_config.json", "w") as f:
    json.dump({"threshold": BEST_THRESHOLD}, f)
print(f"💾 Config saved: detector_config.json  (threshold = {BEST_THRESHOLD:.2f})")

# ── To reload in a fresh session ──────────────────────────────────────────
# from tensorflow.keras.models import load_model
# import json
# model          = load_model("deepfake_detector.keras")
# BEST_THRESHOLD = json.load(open("detector_config.json"))["threshold"]
```

### STEP 9: Predict Function
> Uses the **calibrated threshold** — not a hardcoded 0.5.

```python
def predict_video(video_path):
    """
    Returns (label_string, confidence_percent) for a video file.
    Uses the calibrated BEST_THRESHOLD found on the validation set.
    """
    frames = extract_frames(video_path)

    if len(frames) == 0:
        return "Error: Could not read video ❌", None

    # Normalize the same way training data was normalized
    frames_array = np.array(frames, dtype="float32") / 255.0

    preds    = model.predict(frames_array, verbose=0).flatten()
    avg_pred = float(np.mean(preds))

    # Use calibrated threshold, NOT hardcoded 0.5
    is_real  = avg_pred > BEST_THRESHOLD
    label    = "REAL VIDEO ✅" if is_real else "FAKE VIDEO ❌"

    # Confidence = distance from the threshold, mapped to 0-100%
    if is_real:
        confidence = (avg_pred - BEST_THRESHOLD) / (1.0 - BEST_THRESHOLD)
    else:
        confidence = (BEST_THRESHOLD - avg_pred) / BEST_THRESHOLD

    return label, round(confidence * 100, 1)


# Quick sanity check
test_real = "FaceForensics++/real_original_sequences/000.mp4"
label, conf = predict_video(test_real)
print(f"000.mp4  →  {label}  ({conf}% confident)")
print(f"(Calibrated threshold used: {BEST_THRESHOLD:.2f})")
```

### STEP 10: Interactive UI

```python
import ipywidgets as widgets
from IPython.display import display, Video, clear_output
import tempfile, os

# ── Widgets ────────────────────────────────────────────────────────────────
title_html    = widgets.HTML(
    "<h2 style='color:#2c3e50;margin-bottom:4px'>🎭 Fake Video Detector</h2>"
    f"<p style='color:#888;font-size:13px;margin:0'>Model threshold: {BEST_THRESHOLD:.2f} (auto-calibrated)</p>"
)
upload        = widgets.FileUpload(accept='.mp4', multiple=False, description='📂 Upload')
predict_btn   = widgets.Button(
    description='🔍 Predict',
    button_style='primary',
    layout=widgets.Layout(width='150px', height='40px')
)
clear_btn     = widgets.Button(
    description='🗑️ Clear',
    button_style='warning',
    layout=widgets.Layout(width='100px', height='40px')
)
video_output  = widgets.Output()
result_output = widgets.Output()


def on_predict_clicked(b):
    result_output.clear_output()
    video_output.clear_output()

    if not upload.value:
        with result_output:
            print("⚠️  Please upload a video first.")
        return

    predict_btn.disabled    = True
    predict_btn.description = '⏳ Processing...'

    try:
        # Handle both ipywidgets < 8 (dict) and ≥ 8 (tuple)
        val = upload.value
        if isinstance(val, dict):
            file_info   = list(val.values())[0]
            video_bytes = file_info['content']
            filename    = list(val.keys())[0]
        else:
            file_info   = val[0]
            video_bytes = file_info['content']
            filename    = file_info.get('name', 'video.mp4')

        if isinstance(video_bytes, memoryview):
            video_bytes = bytes(video_bytes)

        # Write to temp file
        tmp = tempfile.NamedTemporaryFile(delete=False, suffix=".mp4")
        tmp.write(video_bytes)
        tmp_path = tmp.name
        tmp.close()

        # Show video
        with video_output:
            print(f"📹 {filename}")
            display(Video(tmp_path, embed=True, width=480))

        with result_output:
            print("⏳ Analyzing video frames...")

        label, confidence = predict_video(tmp_path)

        with result_output:
            clear_output(wait=True)
            if confidence is not None:
                is_real = 'REAL' in label
                color   = '#27ae60' if is_real else '#e74c3c'
                icon    = '✅' if is_real else '❌'
                text    = 'REAL VIDEO' if is_real else 'FAKE VIDEO'
                display(widgets.HTML(f"""
                    <div style='
                        background:{color}18;
                        border:2px solid {color};
                        border-radius:12px;
                        padding:18px 28px;
                        margin-top:12px;
                        font-family:sans-serif;
                    '>
                        <h2 style='color:{color};margin:0 0 6px 0'>{icon}  {text}</h2>
                        <p style='color:#555;margin:0'>Confidence: <b>{confidence}%</b></p>
                        <p style='color:#999;font-size:12px;margin:4px 0 0 0'>
                            Raw score: {label.split()[0]}  |  Threshold: {BEST_THRESHOLD:.2f}
                        </p>
                    </div>
                """))
            else:
                print(label)

        try:
            os.unlink(tmp_path)
        except:
            pass

    except Exception as e:
        with result_output:
            clear_output(wait=True)
            print(f"❌ Error: {e}")

    finally:
        predict_btn.disabled    = False
        predict_btn.description = '🔍 Predict'


def on_clear_clicked(b):
    result_output.clear_output()
    video_output.clear_output()
    upload.value = ()


predict_btn.on_click(on_predict_clicked)
clear_btn.on_click(on_clear_clicked)

display(widgets.VBox([
    title_html,
    upload,
    widgets.HBox([predict_btn, clear_btn]),
    video_output,
    result_output
]))
```
