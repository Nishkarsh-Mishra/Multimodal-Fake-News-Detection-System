# 🎭 Fake Video Detection — FaceForensics++
**Model:** CNN with class-weight balancing | **Dataset:** FaceForensics++ (1000 Real + 5000 Fake)

### STEP 1: Extract Frames

```python
import cv2
import os
import numpy as np

IMG_SIZE = 128
FRAMES_PER_VIDEO = 10  # increased from 3 → better learning per video

def extract_frames(video_path, n_frames=FRAMES_PER_VIDEO):
    """
    Extracts n_frames evenly-spaced frames from a video.
    Evenly spaced sampling gives better coverage than just the first N frames.
    """
    cap = cv2.VideoCapture(video_path)
    if not cap.isOpened():
        return []

    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    if total_frames <= 0:
        cap.release()
        return []

    # Evenly spaced indices across the video
    indices = np.linspace(0, total_frames - 1, n_frames, dtype=int)
    frames = []

    for idx in indices:
        cap.set(cv2.CAP_PROP_POS_FRAMES, idx)
        ret, frame = cap.read()
        if not ret:
            continue
        frame = cv2.resize(frame, (IMG_SIZE, IMG_SIZE))
        frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)  # ✅ Fix: BGR → RGB for correct colors
        frames.append(frame)

    cap.release()
    return frames

print("✅ extract_frames() ready — uses evenly-spaced sampling + BGR→RGB fix")
```

### STEP 2: Load Dataset
**Important:** Dataset has 5000 fake vs 1000 real → class imbalance.  
We'll use `class_weight` during training to handle this properly instead of artificially capping both at 100.

```python
import os
import numpy as np

data = []
labels = []

# ── Limits (adjust based on your RAM) ──────────────────────────────────────
MAX_REAL = 200   # from 1000 available real videos
MAX_FAKE = 200   # from each of 5 fake folders → 1000 total fake videos loaded
                 # keeping 1:1 ratio via class_weight is better than capping
# ───────────────────────────────────────────────────────────────────────────

real_count = 0
fake_count = 0

# ═══════════════════ REAL VIDEOS ═══════════════════
real_path = "FaceForensics++/real_original_sequences"

for video in sorted(os.listdir(real_path)):   # sorted for reproducibility
    if real_count >= MAX_REAL:
        break
    if not video.endswith('.mp4'):
        continue

    video_path = os.path.join(real_path, video)
    frames = extract_frames(video_path)

    for f in frames:
        data.append(f)
        labels.append(1)  # REAL = 1

    real_count += 1
    if real_count % 50 == 0:
        print(f"  Loaded {real_count} real videos")

print(f"✅ Real videos loaded: {real_count}")

# ═══════════════════ FAKE VIDEOS ═══════════════════
fake_path = "FaceForensics++/fake_manipulated_sequences"
fake_folders = sorted(os.listdir(fake_path))   # Deepfakes, Face2Face, FaceShifter, FaceSwap, NeuralTextures
per_folder_limit = MAX_FAKE // len(fake_folders)   # equal share from each manipulation type

for folder in fake_folders:
    folder_path = os.path.join(fake_path, folder)
    if not os.path.isdir(folder_path):
        continue

    folder_count = 0
    for video in sorted(os.listdir(folder_path)):
        if fake_count >= MAX_FAKE or folder_count >= per_folder_limit:
            break
        if not video.endswith('.mp4'):
            continue

        video_path = os.path.join(folder_path, video)
        frames = extract_frames(video_path)

        for f in frames:
            data.append(f)
            labels.append(0)  # FAKE = 0

        fake_count += 1
        folder_count += 1

    print(f"  📁 {folder}: loaded {folder_count} videos")

print(f"✅ Fake videos loaded: {fake_count}")

# ═══════════════════ FINALIZE ═══════════════════
data   = np.array(data,   dtype="float32") / 255.0
labels = np.array(labels, dtype="int32")

print(f"\n📊 Final Data shape : {data.shape}")
print(f"   REAL frames      : {np.sum(labels == 1)}")
print(f"   FAKE frames      : {np.sum(labels == 0)}")
```

### STEP 3: Prepare Data — Train/Val Split + Class Weights

```python
from sklearn.model_selection import train_test_split
from sklearn.utils.class_weight import compute_class_weight

# ── Train / Validation split (80/20) ──────────────────────────────────────
X_train, X_val, y_train, y_val = train_test_split(
    data, labels,
    test_size=0.2,
    random_state=42,
    stratify=labels   # ensures equal class ratio in both splits
)

print(f"Train : {X_train.shape[0]} samples")
print(f"Val   : {X_val.shape[0]} samples")

# ── Class weights (handles 5000 fake vs 1000 real imbalance) ──────────────
class_weights_array = compute_class_weight(
    class_weight='balanced',
    classes=np.unique(y_train),
    y=y_train
)
class_weight_dict = dict(enumerate(class_weights_array))
print(f"\n⚖️  Class weights: {class_weight_dict}")
print("   (Higher weight on minority class = model pays more attention to it)")
```

### STEP 4: Build & Train CNN Model
**Fixes applied:**
- ✅ Added `Dropout` layers to prevent overfitting  
- ✅ Added `BatchNormalization` for stable training  
- ✅ Proper `validation_data` to track real performance  
- ✅ `class_weight` to handle fake/real imbalance  
- ✅ `EarlyStopping` to stop when val_accuracy plateaus  
- ✅ More epochs (15 instead of 3)

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import (
    Conv2D, MaxPooling2D, Flatten, Dense,
    Dropout, BatchNormalization
)
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau

# ── Model Architecture ─────────────────────────────────────────────────────
model = Sequential([
    # Block 1
    Conv2D(32, (3,3), activation='relu', input_shape=(IMG_SIZE, IMG_SIZE, 3), padding='same'),
    BatchNormalization(),
    MaxPooling2D(2, 2),

    # Block 2
    Conv2D(64, (3,3), activation='relu', padding='same'),
    BatchNormalization(),
    MaxPooling2D(2, 2),
    Dropout(0.25),

    # Block 3
    Conv2D(128, (3,3), activation='relu', padding='same'),
    BatchNormalization(),
    MaxPooling2D(2, 2),
    Dropout(0.25),

    # Classifier
    Flatten(),
    Dense(256, activation='relu'),
    BatchNormalization(),
    Dropout(0.5),
    Dense(1, activation='sigmoid')
])

model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=1e-4),
    loss='binary_crossentropy',
    metrics=['accuracy', tf.keras.metrics.AUC(name='auc')]
)

model.summary()

# ── Callbacks ──────────────────────────────────────────────────────────────
callbacks = [
    EarlyStopping(
        monitor='val_accuracy',
        patience=4,
        restore_best_weights=True,
        verbose=1
    ),
    ReduceLROnPlateau(
        monitor='val_loss',
        factor=0.5,
        patience=2,
        verbose=1
    )
]

# ── Train ──────────────────────────────────────────────────────────────────
history = model.fit(
    X_train, y_train,
    epochs=15,
    batch_size=32,
    validation_data=(X_val, y_val),
    class_weight=class_weight_dict,
    callbacks=callbacks
)

# ── Final Validation Score ─────────────────────────────────────────────────
val_loss, val_acc, val_auc = model.evaluate(X_val, y_val, verbose=0)
print(f"\n🏆 Final Val Accuracy : {val_acc*100:.2f}%")
print(f"   Final Val AUC      : {val_auc:.4f}")
```

### STEP 5: Plot Training History

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Accuracy plot
axes[0].plot(history.history['accuracy'],     label='Train Accuracy', marker='o')
axes[0].plot(history.history['val_accuracy'], label='Val Accuracy',   marker='o')
axes[0].set_title('Model Accuracy')
axes[0].set_xlabel('Epoch')
axes[0].set_ylabel('Accuracy')
axes[0].legend()
axes[0].grid(True)

# Loss plot
axes[1].plot(history.history['loss'],     label='Train Loss', marker='o')
axes[1].plot(history.history['val_loss'], label='Val Loss',   marker='o')
axes[1].set_title('Model Loss')
axes[1].set_xlabel('Epoch')
axes[1].set_ylabel('Loss')
axes[1].legend()
axes[1].grid(True)

plt.suptitle('Training History', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('training_history.png', dpi=100, bbox_inches='tight')
plt.show()
print("💾 Saved: training_history.png")
```

### STEP 6: Confusion Matrix + Classification Report

```python
from sklearn.metrics import confusion_matrix, classification_report, ConfusionMatrixDisplay

y_pred_prob = model.predict(X_val, verbose=0)
y_pred      = (y_pred_prob > 0.5).astype(int).flatten()

cm = confusion_matrix(y_val, y_pred)

fig, ax = plt.subplots(figsize=(6, 5))
disp = ConfusionMatrixDisplay(
    confusion_matrix=cm,
    display_labels=['FAKE', 'REAL']
)
disp.plot(ax=ax, colorbar=False, cmap='Blues')
ax.set_title('Confusion Matrix (Validation Set)', fontweight='bold')
plt.tight_layout()
plt.savefig('confusion_matrix.png', dpi=100, bbox_inches='tight')
plt.show()

print("\n📋 Classification Report:")
print(classification_report(y_val, y_pred, target_names=['FAKE', 'REAL']))
```

### STEP 8: Predict Function

```python
def predict_video(video_path, threshold=0.5):
    """
    Returns prediction label + confidence score for a given video file.
    """
    frames = extract_frames(video_path)

    if len(frames) == 0:
        return "Error: Could not read video ❌", None

    frames_array = np.array(frames, dtype="float32") / 255.0
    preds        = model.predict(frames_array, verbose=0)
    avg_pred     = float(np.mean(preds))

    if avg_pred > threshold:
        label = "REAL VIDEO ✅"
    else:
        label = "FAKE VIDEO ❌"

    confidence = avg_pred if avg_pred > threshold else (1 - avg_pred)
    return label, round(confidence * 100, 2)


# Quick test
test_path = "FaceForensics++/real_original_sequences/000.mp4"
label, confidence = predict_video(test_path)
print(f"Result     : {label}")
print(f"Confidence : {confidence}%")
```

### STEP 9: Interactive UI (Fixed)

**Bugs fixed:**
- ✅ `upload.value` was a tuple in newer ipywidgets — now handled correctly  
- ✅ Added confidence score display  
- ✅ Added progress bar while processing  
- ✅ Result shown with color (green = real, red = fake)  
- ✅ Button stays disabled while processing (prevents double clicks)  

```python
import ipywidgets as widgets
from IPython.display import display, Video, clear_output
import tempfile
import os

# ── Widgets ────────────────────────────────────────────────────────────────
title_html    = widgets.HTML("<h2 style='color:#2c3e50'>🎭 Fake Video Detector</h2>")
upload        = widgets.FileUpload(accept='.mp4', multiple=False, description='📂 Upload')
predict_btn   = widgets.Button(
    description='🔍 Predict',
    button_style='primary',
    layout=widgets.Layout(width='150px', height='40px')
)
clear_btn     = widgets.Button(
    description='🗑️ Clear',
    button_style='warning',
    layout=widgets.Layout(width='100px', height='40px')
)
video_output  = widgets.Output()
result_output = widgets.Output()


def on_predict_clicked(b):
    result_output.clear_output()
    video_output.clear_output()

    # ── FIX: handle both old dict style and new tuple style of ipywidgets ──
    if not upload.value:
        with result_output:
            print("⚠️  Please upload a video first.")
        return

    predict_btn.disabled = True
    predict_btn.description = '⏳ Processing...'

    try:
        # ipywidgets ≥ 8.x  → upload.value is a tuple of dicts
        # ipywidgets < 8.x  → upload.value is a dict keyed by filename
        val = upload.value
        if isinstance(val, dict):
            file_info   = list(val.values())[0]
            video_bytes = file_info['content']
            filename    = list(val.keys())[0]
        else:
            # tuple / list of FileInfo objects (ipywidgets ≥ 8)
            file_info   = val[0]
            video_bytes = file_info['content']  # memoryview or bytes
            filename    = file_info.get('name', 'video.mp4')

        # Convert memoryview → bytes if needed
        if isinstance(video_bytes, memoryview):
            video_bytes = bytes(video_bytes)

        # Write to temp file
        temp_file = tempfile.NamedTemporaryFile(delete=False, suffix=".mp4")
        temp_file.write(video_bytes)
        temp_path = temp_file.name
        temp_file.close()

        # Show the uploaded video
        with video_output:
            print(f"📹 File: {filename}")
            display(Video(temp_path, embed=True, width=480))

        # Show loading message
        with result_output:
            print("⏳ Analyzing video frames...")

        # Run prediction
        label, confidence = predict_video(temp_path)

        # Display result with color
        with result_output:
            clear_output(wait=True)
            if confidence is not None:
                color  = '#27ae60' if 'REAL' in label else '#e74c3c'
                icon   = '✅' if 'REAL' in label else '❌'
                text   = 'REAL VIDEO' if 'REAL' in label else 'FAKE VIDEO'
                display(widgets.HTML(f"""
                    <div style='
                        background:{color}22;
                        border:2px solid {color};
                        border-radius:10px;
                        padding:16px 24px;
                        margin-top:10px;
                        font-family:sans-serif;
                    '>
                        <h2 style='color:{color};margin:0'>{icon} {text}</h2>
                        <p style='color:#555;margin:6px 0 0 0'>Confidence: <b>{confidence}%</b></p>
                    </div>
                """))
            else:
                print(label)

        # Clean up temp file
        try:
            os.unlink(temp_path)
        except:
            pass

    except Exception as e:
        with result_output:
            clear_output(wait=True)
            print(f"❌ Error: {str(e)}")

    finally:
        predict_btn.disabled    = False
        predict_btn.description = '🔍 Predict'


def on_clear_clicked(b):
    result_output.clear_output()
    video_output.clear_output()
    # Reset the upload widget
    upload.value = ()


predict_btn.on_click(on_predict_clicked)
clear_btn.on_click(on_clear_clicked)

# ── Layout ─────────────────────────────────────────────────────────────────
buttons = widgets.HBox([predict_btn, clear_btn])
ui = widgets.VBox([
    title_html,
    upload,
    buttons,
    video_output,
    result_output
])

display(ui)
```
